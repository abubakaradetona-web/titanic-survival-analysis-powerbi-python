```python
import numpy as np
df['Cabin'] = df['Cabin'].replace('\\N', np.nan)
print(df['Cabin'].unique())
print(df['Embarked'].unique())
df.head()
```

```python
# Replace '\N' strings with NaN
df['Embarked'] = df['Embarked'].replace('\\N', np.nan)

# Replace '\N' strings with NaN
df['Age'] = df['Age'].replace('\\N', np.nan)

# Convert to numeric, coercing errors to NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Fill NaN values with the mean of the 'Age' column
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Convert to integer type
df['Age'] = df['Age'].astype(int)

print(df['Age'].dtype)
df.head()
```

```python
## Renamed all columns to lowercase and removed all space in the column name.

df.columns = df.columns.str.lower().str.replace(' ', '')
df.head()
```

```python
## Identified all strings column in my table

object_columns = df.select_dtypes(include='object').columns.tolist()
print("Columns with object data type:", object_columns)
```

```python
## converted all Strings column text value to lowercase and removed all space
for col in object_columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str).str.strip().str.lower()

df.head()
```

```python
#created a new deck column by exctracting the first initials from the cabin column

df['deck'] = df['cabin'].apply(lambda x: x[0] if pd.notna(x) and x != 'nan' else np.nan)
df.head()
```

```python
## Replaced all Nan in Embarked colum with the mode of the Embarked column

df['embarked'] = df['embarked'].replace('nan', np.nan)
mode_embarked = df['embarked'].mode()[0]
df['embarked'] = df['embarked'].fillna(mode_embarked)
df.head()
```

```python
## created  a new agegroup column

bins = [0, 12, 20, 60, df['age'].max()]
labels = ['Child', 'Teen', 'Adult', 'Senior']
df['agegroup'] = pd.cut(df['age'], bins=bins, labels=labels, right=True)

df.head()
```

```python
## created the passengertitle column from the name column by the regular expression module
import re

def extract_title(name):
    title_search = re.search(r'([a-z]+)\.', name)
    if title_search:
        title = title_search.group(1)
        if title in ['mr', 'mrs', 'miss']:
            return title
    return 'other'

df['passengertitle'] = df['name'].apply(extract_title)
df.head()
```

```python
df['familysize'] = df['sibsp'] + df['parch'] + 1
df.head()
```

```python
# Export the DataFrame to a CSV file
output_csv_path = 'cleaned_titanic_data.csv'
df.to_csv(output_csv_path, index=False)

print(f"Data exported to {output_csv_path}")
```
